```python
import pandas as pd
data = pd.read_csv('creditcard_2023.csv')
print(data.head())
print(data.info())
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data =pd.read_csv('creditcard_2023.csv')
plt.figure(figsize=(10,6))
sns.lineplot(x='Class',y='Amount',data=data)
plt.title('Credit Card Fraud Detection')
plt.xlabel('Class')
plt.ylabel('Amount')
plt.show()
```

```python
f1, (ax1, ax2) = plt.subplots(2, 1, sharex =True)
f1.suptitle('Class wise amount per transaction')
bins = 50
ax1.hist(f.Amount, bins = bins)
ax1.set_title('Fraud')

ax2.hist(n.Amount, bins = bins)
ax2.set_title('Normal')

plt.xlabel('Amount')
plt.ylabel('No. of transactions')
plt.xlim((0,20000))
plt.yscale('log')
plt.show()
```

```python
count_classes = pd.value_counts(data['Class'], sort = True)
count_classes.plot(kind = 'bar')
plt.title('Frequency of class')
plt.xlabel('Class')
plt.ylabel('Frequency')
```
